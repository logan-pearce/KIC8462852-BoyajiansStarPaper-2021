# KIC 8462852 B Astrometry

This notebook takes in the summary files that are aggregates of the MCMC chains used to fit the PSF of the star and companion in each image.  The format of these text files is: <br>
Image filename, Sep mean (mas), sep error (mas), pa mean (deg), pa err (deg), SNR primary, SNR secondary, DeltaMag, FWHM mean, FWHM err, Mean Gelman-Rubin stat for chains, GR stat std dev

Each step in the MCMC chain was corrected for detector distortion and atmopheric correction and burn in before computing the means in this file.

This notebook takes the image statistics in an epoch and computes epoch means, and produces plots in Fig 3, 6, and 7

```python
############################################ 2014 #########################################################

pixscale = 9.952 #mas/pix
# Open MCMC chain files:
results = "2014_epoch_pasep_B"
pos = np.loadtxt(open(results,"rb")\
                 ,delimiter=",",usecols=(1,2,3,4,5,6,7,8,9))
filename = np.loadtxt(open(results,"rb"),\
                       dtype=str,delimiter=",",usecols=(0,))

# Parse all the data into arrays:
r = pos[:,0]  #Distance in arcseconds
r_std = pos[:,1] #std dev of sep
t_deg = pos[:,2]-180.  #Angle in degrees
t_std_deg = pos[:,3] # Angle std dev

# obs date in decimal year:
date = 2014.7890411

# Convert angle to mas - rho*DeltaTheta, where Theta is angle from due east:
t = r*np.tan(np.radians(t_deg-90.))
# Convert angle error to mas:
t_std = r*np.tan(np.radians(t_std_deg))

##### Mean values for separation and position angle:
r_weights = 1./(r_std**2)
r_mean = np.sum(r_weights*r)/np.sum(r_weights)
t_weights = 1./(t_std**2)
t_mean = np.sum(t_weights*t)/np.sum(t_weights)
t_weights_deg = 1./(t_std_deg**2)
t_mean_deg = np.sum(t_weights_deg*t_deg)/np.sum(t_weights_deg)

##### Array of uncertanties in each image:
rerr_thick = np.sqrt((1./np.sum(r_weights)))
terr_thick = np.sqrt((1./np.sum(t_weights)))

##### Systematic Uncertainties:
# Source for weighted means: https://ned.ipac.caltech.edu/level5/Leo/Stats4_5.html

### Dither 1: ###
## Separation:
# Add in Sqrt(2) uncert to sep and pa for differential positional uncertainty for each dither 
# from comparing comp position to star position:
uncerts1 = np.sqrt((np.sqrt((1./np.sum(r_weights[0]))))**2 + (np.sqrt(2)**2))
## Position angle:
uncertpa1 = np.sqrt((np.sqrt((1./np.sum(t_weights[0]))))**2 + (np.sqrt(2)**2))

### Dither 2: ###
## Separation:
uncerts2 = np.sqrt((np.sqrt((1./np.sum(r_weights[1:]))))**2 + (np.sqrt(2)**2))
## Position angle:
uncertpa2 = np.sqrt((np.sqrt((1./np.sum(t_weights[1:]))))**2 + (np.sqrt(2)**2))

#### Weighted mean of the uncertainty from the two dithers:
# The uncert in each dither times the number of images in each dither, over the total number of images:
#uncerts = 1/np.sqrt((1./np.sum([uncerts1**2,uncerts2**2])))
#uncertpa = 1/np.sqrt((1./np.sum([uncertpa1**2,uncertpa2**2])))
N_dithers = 2
uncerts = np.mean([uncerts1,uncerts2])/np.sqrt(N_dithers)
uncertpa = np.mean([uncertpa1,uncertpa2])/np.sqrt(N_dithers)

#### Add in Pix scale and orientation errors
## Add pix scale uncert of 0.002 mas/pix to separation
# Convert sep to pixels:
rpix = r/pixscale #radius in pixels
pixerr = rpix*0.002 #pix * mas/pix = mas
# Add in quadrature to sep uncert:
uncerts = np.mean(np.sqrt(uncerts**2+pixerr**2))

## Add orientation correction uncert to PA:
# convert orientation uncertainty to mas:
orerr = 0.009 #deg
orerr = np.radians(orerr)
orerr = r*np.tan(orerr)
# Add in quad with pa uncert:
uncertpa = np.mean(np.sqrt(uncertpa**2+orerr**2))

# Convert PA uncert to degrees:
terr_deg = np.degrees(np.arctan(uncertpa/np.mean(r_mean)))

################################################ Final output: ################################################
r1,t1,t1deg = r,t,t_deg
r1mean,t1mean,t1meandeg = r_mean,t_mean,t_mean_deg
t1err_deg = terr_deg
r1_std,t1_std_deg = r_std,t_std_deg
r1err_thick,t1err_thick = np.std(r),np.std(t)

r1err_thin,t1err_thin = uncerts,uncertpa
r1err_ind = np.median(r_std)
t1err_ind = np.median(t_std)
d1 = date
deltamag1 = pos[:,6]

# Overall means:
r2014, t2014, tdeg2014 = r1,t1,t1deg
r2014_mean,t2014_mean,t2014_meandeg = r1mean,t1mean,t1meandeg
r2014_err_thick,t2014_err_thick = r1err_thick,t1err_thick
r2014_err_std,t2014_err_std = np.std(r1),np.std(t1)
r2014_err_thin,t2014_err_thin = r1err_thin,t1err_thin

# Separate by filter:
sep_Kfilter_2014 = np.array([np.sum(r_weights[0:10]*r1[0:10])/np.sum(r_weights[0:10]) , np.std(r1[0:10]) ])
sep_Jfilter_2014 = np.array([np.sum(r_weights[10:20]*r1[10:20])/np.sum(r_weights[10:20]) , np.std(r1[10:20]) ])
sep_Hfilter_2014 = np.array([np.sum(r_weights[20:]*r1[20:])/np.sum(r_weights[20:]) , np.std(r1[20:]) ])

pa_Kfilter_2014 = np.array([np.sum(t_weights[0:10]*t1deg[0:10])/np.sum(t_weights[0:10]) , np.std(t1deg[0:10]) ])
pa_Jfilter_2014 = np.array([np.sum(t_weights[10:20]*t1deg[10:20])/np.sum(t_weights[10:20]) , np.std(t1deg[0:10]) ])
pa_Hfilter_2014 = np.array([np.sum(t_weights[20:]*t1deg[20:])/np.sum(t_weights[20:]) , np.std(t1deg[20:]) ])
```

### Testing background vs comoving hypothesis
Using the covariance matrix described in the paper, Sec 2.2

$$\chi^2 = \textbf{r}^T \, \textbf{C}^{-1} \, \textbf{r}$$
where C = covariance matrix, r = residuals of data from model under test.


```python
# Compute the residuals from the model (zero movement)
data = np.array([r2014_mean,r2016_1_mean,r2019_1_mean,r2016_2_mean,r2019_2_mean,\
                 t2014_mean,t2016_1_mean,t2019_1_mean,t2016_2_mean,t2019_2_mean])

ref_rmean = np.average([r2014_mean,r2016_1_mean,r2016_2_mean,r2019_1_mean,\
                                      r2019_2_mean], 
                                weights = np.array([r2014_err_thin,r2016_1_err_thin,rerr2016_2_thick,\
                                                  r2019_1_err_thin,rerr2019_2_thick])**(-1))

ref_tmean = np.average([t2014_mean,t2016_1_mean,t2016_2_mean,t2019_1_mean,\
                                      t2019_2_mean], 
                                        weights = 1/np.array([t2014_err_thin,t2016_1_err_thin,terr2016_2_thick,\
                                                  t2019_1_err_thin,terr2019_2_thick]))

# model: every point would be the weighted average of all:
model = np.array([ref_rmean,ref_rmean,ref_rmean,ref_rmean,ref_rmean,\
                  ref_tmean,ref_tmean,ref_tmean,ref_tmean,ref_tmean])

residuals = data - model
```

```python
# Covariance matrix:
cov = np.zeros((10,10))
sigmas = np.array([r2014_err_thick,r2016_1_err_thick,r2019_1_err_thick,r2016_2_err_thick,r2019_2_err_thick,\
                  t2014_err_thick,t2016_1_err_thick,t2019_1_err_thick,t2016_2_err_thick,t2019_2_err_thick])
sigmas_std = np.array([r2014_err_std,r2016_1_err_std,r2019_1_err_std,r2016_2_err_std,r2019_2_err_std,\
                  t2014_err_std,t2016_1_err_std,t2019_1_err_std,t2016_2_err_std,t2019_2_err_std])
# populate diagonals:
value = 2
for i in range(len(sigmas)):
    cov[i,i] = sigmas_std[i]**2 + value
    
# off diagonals:
# r2016-1/r2019-1:
cov[2,1] = value
# r2019-1/r2016-1:
cov[1,2] = value
# r2016-2/r2019-2:
cov[4,3] = value
# r2019-2/r2016-2:
cov[3,4] = value
# t2016-1/t2019-1:
cov[7,6] = value
# t2019-1/t2016-1:
cov[6,7] = value
# t2016-2/t2019-2:
cov[9,8] = value
# t2019-2/t2016-2:
cov[8,9] = value

# Compute Chi^2
Weight = np.linalg.inv(cov)
Wdotr = np.matmul(Weight,residuals)
Chi2 = np.matmul(residuals.T,Wdotr)
print('Chi2:',Chi2)
```
